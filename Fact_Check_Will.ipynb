# Fact checking discipline analysis below

```python
discipline = pd.read_csv("/Users/josemartinez/Desktop/Code/Discipline_Data/cleaned_discipline_final.csv")
```

```python
discipline = discipline[
    (discipline.clean_race_ethnicity!='unknown') & (discipline.roster_name_match!='no match') & ~(discipline.clean_race_ethnicity.isna())
].copy()
```

```python
discipline.columns = [
    'control_number',
    'final_disposition_date',
    'final_disposition',
    'final_number_of_days',
    'last_name',
    'first_name',
    'allegation',
    'complaintant_gender',
    'complaintant_race',
    'dirty_full_name',
    'clean_discipline_name',
    'clean_roster_name',
    'artifact_roster_name',
    'officer_gender',
    'officer_race',
]
```

```python
discipline
```

```python
dis_by_gender = discipline.groupby(['officer_gender'])['officer_gender'].count().to_frame(name='amount').reset_index()
```

```python
dis_by_gender
```

```python
dis_by_gender['% of dept'] = (dis_by_gender['amount']/dis_by_gender['amount'].sum())*100
```

## ^Matches Will's findings about racial makeup of discipline data.

```python
dis_by_race = discipline.groupby(['officer_race'])['officer_race'].count().to_frame(name='amount').reset_index()
```

```python
dis_by_race = dis_by_race[dis_by_race['officer_race'] != 'unknown']
```

## Matches Will's findings about racial share of allegations.

```python
race_dispo = discipline.groupby(['officer_race','final_disposition'])['final_disposition'].count().to_frame(name='amounts').reset_index()
```

```python
severe_dispo_race = race_dispo.loc[race_dispo['final_disposition'].str.contains('SUSPEN|DEMOT|TERMIN')].groupby(['officer_race'])['amounts'].sum().to_frame(name='amounts').reset_index()
```

```python
all_dispo_race = discipline.groupby(['officer_race','final_disposition'])['final_disposition'].count().to_frame(name='amount').reset_index().groupby(['officer_race'])['amount'].sum().to_frame(name='amounts').reset_index()
```

```python
severe_dispo_race['% of total'] = (severe_dispo_race['amounts']/all_dispo_race['amounts'])*100
```

```python
severe_dispo_race
```

## ^Matches Will's findings about racial makeup of severe dispositions.

```python
race_gender_dispo = discipline.groupby(['officer_race','officer_gender','final_disposition'])['final_disposition'].count().to_frame(name='amounts').reset_index()
```

```python
severe_dispo_race_gender = race_gender_dispo.loc[race_gender_dispo['final_disposition'].str.contains('SUSPEN|DEMOT|TERMIN')].groupby(['officer_race','officer_gender'])['amounts'].sum().to_frame(name='amounts').reset_index()
```

```python
all_dispo_race_gender = discipline.groupby(['officer_race','officer_gender','final_disposition'])['final_disposition'].count().to_frame(name='amount').reset_index().groupby(['officer_race','officer_gender'])['amount'].sum().to_frame(name='amounts').reset_index()
```

```python
all_dispo_race_gender = all_dispo_race_gender.loc[all_dispo_race_gender['amounts']!=2].reset_index()
```

```python
severe_dispo_race_gender['% of total'] = (severe_dispo_race_gender['amounts']/all_dispo_race_gender['amounts'])*100
```
